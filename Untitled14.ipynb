```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('C:/Users/SKG/Downloads/USA_Housing.csv')
data = pd.DataFrame(dataset)
data = dataset.drop('Address',axis = 1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data.drop('Price',axis = 1))

x1 = scaled_features[:,0]
x2 = scaled_features[:,1]
x3 = scaled_features[:,2]
x4 = scaled_features[:,3]
x5 = scaled_features[:,4]
y = np.array(data['Price'])

a = 0.09
m = 5000

def cost_function(t0,t1,t2,t3,t4,t5,m=5000):
    s = 0
    for i in range(m):
        s += (t0 + t1*x1[i] + t2*x2[i] + t3*x3[i] + t4*x4[1] + t5*x5[i] - y[i])**2
    return s/(2*m)

def gradient_decent_theta(t0,t1,t2,t3,t4,t5,a):
    g0 = t0 - (a * ((m*t0 + t1*sum(x1) + t2*sum(x2) + t3*sum(x3) + t4*sum(x4) + t5*sum(x5) - sum(y))/m))
    g1 = t1 - (a * ((t0*sum(x1) + t1*sum((x1*x1)) + t2*sum(x2*x1) + t3*sum(x3*x1) + t4*sum(x4*x1) + t5*sum(x5*x1) -sum(y*x1))/m))
    g2 = t2 - (a * ((t0*sum(x2) + t1*sum((x1*x2)) + t2*sum(x2*x2) + t3*sum(x3*x2) + t4*sum(x4*x2) + t5*sum(x5*x2) -sum(y*x2))/m))
    g3 = t3 - (a * ((t0*sum(x3) + t1*sum((x1*x3)) + t2*sum(x2*x3) + t3*sum(x3*x3) + t4*sum(x4*x3) + t5*sum(x5*x3) -sum(y*x3))/m))
    g4 = t4 - (a * ((t0*sum(x4) + t1*sum((x1*x4)) + t2*sum(x2*x4) + t3*sum(x3*x4) + t4*sum(x4*x4) + t5*sum(x5*x4) -sum(y*x4))/m))
    g5 = t5 - (a * ((t0*sum(x5) + t1*sum((x1*x5)) + t2*sum(x2*x5) + t3*sum(x3*x5) + t4*sum(x4*x5) + t5*sum(x5*x5) -sum(y*x5))/m))
    return (g0,g1,g2,g3,g4,g5)


number_of_itteration = 1
itteration  = list()
cost = list()
t0 = t1 = t2 = t3 = t4 = t5 = 0
itteration.append(number_of_itteration)
cost.append(cost_function(t0,t1,t2,t3,t4,t5))
while(1):
    c1 = 0
    c2 = 0
    c1 = cost_function(t0,t1,t2,t3,t4,t5)
    (t0,t1,t2,t3,t4,t5) = gradient_decent_theta(t0,t1,t2,t3,t4,t5,a)
    c2 = cost_function(t0,t1,t2,t3,t4,t5)
    if c2 > c1:
        print(t0,t1,t2,t3,t4,t5,cost_function(t0,t1,t2,t3,t4,t5))
        break
    else:
        number_of_itteration += 1
        itteration.append(number_of_itteration)
        cost.append(c2)
```

```python
plt.figure(figsize = (20,12))
plt.plot(itteration,cost,label = 'cost vs iteration')
plt.legend(loc = 0)
plt.xlabel('Number of iterations')
plt.ylabel('cost function')
```

```python
y_predicted = t0*np.ones(m) + t1*x1 + t2*x2 + t3*x3 + t4*x4 + t5*x5
```

```python
plt.figure(figsize = (20,12))
plt.scatter(y,y_predicted,label = 'actual price vs predicted price')
plt.legend(loc = 0)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
```
